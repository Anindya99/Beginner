```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

def money(s):
    s=str(s)
    if s[0]==('€'):
        s = s[1:]
    multiplier = 0
    if s[len(s)-1]==('M'):
        s = s[:-1]
        multiplier = pow(10,6)
    if s[len(s)-1]==('B'):
        s = s[:-1]
        multiplier = pow(10,9)
    if s[len(s)-1]==('K'):
        s = s[:-1]
        multiplier = pow(10,3)
    f = float(s)
    f = f * multiplier
    return f

def height(h):
    h=str(h)
    h2=-1
    for c in h:
        h2+=1
        if c==("'"):
            h=h.replace(h[h2],'.',1)
            break;
    h=float(h)
    return h;

def weight(w):
    w=str(w)
    h2=-1
    for c in w:
        h2+=1
        if c==("l"):
            w=w[:h2]
            break;
    w=float(w)
    return w;

df = pd.read_csv("/home/nbuser/library/dataFifa19.csv",index_col='ID')
df.head()
df=df.drop('Unnamed: 0',axis=1)
df.head()
df['Value']=df['Value'].apply(money)
df['Wage']=df['Wage'].apply(money)
df['Release Clause']=df['Release Clause'].apply(money)
df.head(10)
df['Height']=df['Height'].apply(height)
df['Height'].tail(10)
df['Weight']=df['Weight'].apply(weight)
df.head()
```
