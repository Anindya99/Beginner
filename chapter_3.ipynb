# Multiple Linear Regression

```r
lm.fit=lm(medv~lstat+age,data=Boston)
## obtaining a confidence interval for the coefficient estimates
summary(lm.fit)
```

```r
library(MASS)
library(ISLR)
lm.fit=lm(medv~.,data=Boston)
summary(lm.fit)
```

```r
# Removing the variable 'age' from prediction as it doesn't have significant contribution in the prediction, as evident
# from the above summary.
lm.fit1=update(lm.fit,~.-age)
```

# Using age and lstat both the variables together to predict,by using interpretation.

```r
summary(lm(medv~lstat*age,data=Boston))
```

```r
lm.fit2=lm((medv~lstat+I(lstat^2)))
summary(lm.fit2)
```

```r
lm.fit=lm(medv~lstat,data=Boston)
anova(lm.fit2,lm.fit)
```

```r

plot(lm.fit2)
```

```r
lm.fit5=lm(medv~poly(lstat,5))
summary(lm.fit5)
```

```r
lm.fitlog=lm(medv~log(rm))
summary(lm.fitlog)
```
